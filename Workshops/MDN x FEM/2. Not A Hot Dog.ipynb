# DataBunch

Now that we have images in `hot_dog` and `not_hot_dog`, we want to be able to load an image from the folder into numbers that we will store in a variable. We also want to load training and testing images separately, otherwise we won't be able to test the model.

fastai provides us an easy way of doing that. Here is how the code works: 
- `ImageList.from_folder` tells fastAI that the image to look for is in 'path'.
- `random_split_by_pct()` tells the model to generate a training/test set randomly. By default the split is 80/20. Remember: train/test split is very important!
- `label_from_folder()` tells the model that the label of image is determined by the name of the folder it's in, which in our case is `hot_dog` and `not_hot_dog`.
- `.transform(size=224)` "transforms" the images (by turning them into numbers) and resizes them to a size of 224 pixels. This standardizes the size of all images going into the network. **This is generally the case with most ML model training.**
- `.databunch()` takes the ImageList and generates a DataBunch, which can be passed into a fastAI learner object.

```
# this will ask you to sign in and give permission. Don't worry, no malware :)
from google.colab import drive
drive.mount('/content/drive')
```

```
path = Path('hotdog-not-hotdog-dataset')

data_bunch = (
    fastai.vision.ImageList.from_folder(path=path)
    .random_split_by_pct()
    .label_from_folder()
    .transform(size=224)
    .databunch()
)
```

## DataBunch Inspect

Let's see some of our data. First, let's print out what our `data_bunch` actually contains.

```
print("Training set images: \n", data_bunch.train_ds.x)
print("="*40)
print("Training set labels: \n", data_bunch.train_ds.y)
print('\n') # print empty lines
print("Testing set images: \n", data_bunch.valid_ds.x)
print("="*40)
print("Testing set labels: \n", data_bunch.valid_ds.y)
```

Let's see what kind of classes we have and how many of them there are.

```
print("Number of training set images:", len(data_bunch.train_ds))
print("Number of testing set images:", len(data_bunch.valid_ds))
print("Unique classes:", data_bunch.classes)
```

Let's see some of those images

# Model

We will use a neural network architecture known as ResNet50 (hence `base_arch=fastai.vision.models.resnet50`. It is essentially a neural network, but with some modifications which make it much more powerful at image recognition, although we will not go into these.

Note that we are providing `pretrained=True` to the model. This is telling fastai to fetch a trained model for us, so we don't have to train from the ground up. Most interestingly, the dataset this model is trained on has nothing to do with hotdogs! The model has been trained on a much larger dataset to tell apart a lot of different things (including hotdog). We are relying on all the rules the model has learned for that task to speed up our learning on this task. This is a powerful technique in Deep Learning, called transfer learning. 

If you set this parameter to False, things will take much longer!

```
learner = fastai.vision.cnn_learner(
    data=data_bunch, 
    base_arch=fastai.vision.models.resnet50,
    pretrained=True, 
    metrics=fastai.vision.accuracy # this is a function which computes accuracy
)
```

## Saving Model Weights
- Saving model weights is an important and good practice when working with ML models. It is often the case that there is a bug somewhere which crashes your code halfway through training, which means you have already put it days of computation. It is very painful to have to wait that long again.
- By saving the weights, we can always just load it back again to resume training to deal with such disruptions.
- `.save(<file>)` saves the weights of the model into `<file>`. This file will sit around on your storage.
- <odel weights can be simply reloaded with the syntax **`learner.load(<file>)`**

```
learner.save('pretrained')
```

# Training

## Let The Training Begin
- `learner.fit_one_cycle` cycles through the dataset once. We call this an **epoch**
- In a single training step, images are drawn from our dataset, fed through the model, and the model change its weights based on how well it does. Once we have drawn all images from the dataset, we have finished an epoch. We usually want to train for multiple epochs, where we recycle the same images from our dataset.
- The number of epochs to train for can vary substantially due to model, pretrained weights, data, classes, and so forth. It is determined largely by experimentation.

The first argument we provide to `fit_one_cycle`, 5, is the number of epochs. It will take ~6 minutes to train 5 epochs on GPU.

Note we are also providing the argument `1e-04` which is equal to 0.0001. This is the **learning_rate**. You will know more about this in the next workshop. Please ask us if you are curious, but for now, just think of it as a magic number :)

```
learner.fit_one_cycle(5, 1e-04)
```

Let's see some predictions

```
learner.show_results()
```

```
learner.save('first-model')
```

## It's not done till it's tested - Model Validation
- An important part of validating ML models is to generate metrics for you to see how well the model is performing (on the test set!!!!). fastai will do this for you.
- Confusion matrix is a great way to understand the model errors. If the name confuses you, actually seeing the matrix will make things clear. Check that the numbers in the confusion matrix adds up.

```
# plot confusion matrix
interp = fastai.vision.ClassificationInterpretation.from_learner(learner)
interp.plot_confusion_matrix()
```

## Lets look at the areas our model sucks
- the plot_top_losses function allows you to look at on which images does the model perform the worst on. 
- If the model consistently performs bad on class 'A', it is a hint that something is wrong with the model, and calls for more investigation.
- A heatmap is a really important tool is determine what parts of an image does the model use to make decisions.

```
# prediction, actual, loss, and probability
interp.plot_top_losses(4, heatmap=True)
```
